### Fetching repos data

```python
def get_repo_data(owner, repo_name, tech):
    url = f'https://api.github.com/repos/{owner}/{repo_name}'
    response = requests.get(url, headers=HEADERS)
    if response.status_code == 200:
        repo = response.json()
        releases_url = repo['releases_url'].replace('{/id}', '')
        releases_response = requests.get(releases_url, headers=HEADERS)
        if releases_response.status_code == 200:
            releases_count = len(releases_response.json())
        else:
            releases_count = 0
        data = {
            'name': repo['name'],
            'full_name': repo['full_name'],
            'description': repo['description'],
            'search_key': tech,
            'language': repo['language'],
            'key_words': repo['topics'],
            'stars': repo['stargazers_count'],
            'forks': repo['forks_count'],
            'watchers': repo['watchers_count'],
            'open_issues': repo['open_issues_count'],
            'created_at': repo['created_at'],
            'updated_at': repo['updated_at'],
            'pushed_at': repo['pushed_at'],
            'releases': releases_count
        }
        return data
    else:
        print(f"Failed to fetch data for {owner}/{repo_name}: {response.status_code}")
        return None
```

### search for repos

```python
def search_repos(query, per_page=10000):
    url = f'https://api.github.com/search/repositories?q={query}&sort=stars&order=desc&per_page={per_page}'
    response = requests.get(url, headers=HEADERS)
    if response.status_code == 200:
        return response.json()['items']
    else:
        print(f"Failed to search repositories: {response.status_code}")
        return []
```

### main

```python
import requests
import pandas as pd
import time

# Replace 'your_access_token_here' with your actual GitHub personal access token
ACCESS_TOKEN = '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'
HEADERS = {'Authorization': f'token {ACCESS_TOKEN}'}

def get_repo_data(owner, repo_name, tech):
    url = f'https://api.github.com/repos/{owner}/{repo_name}'
    try:
        response = requests.get(url, headers=HEADERS)
        response.raise_for_status()  # Raise exception for 4xx and 5xx status codes
        repo = response.json()
        releases_url = repo['releases_url'].replace('{/id}', '')
        releases_response = requests.get(releases_url, headers=HEADERS)
        releases_response.raise_for_status()
        releases_count = len(releases_response.json())
        data = {
            'name': repo['name'],
            'full_name': repo['full_name'],
            'description': repo['description'],
            'search_key': tech,
            'language': repo['language'],
            'key_words': repo['topics'],
            'stars': repo['stargazers_count'],
            'forks': repo['forks_count'],
            'watchers': repo['watchers_count'],
            'open_issues': repo['open_issues_count'],
            'created_at': repo['created_at'],
            'updated_at': repo['updated_at'],
            'pushed_at': repo['pushed_at'],
            'releases': releases_count
        }
        return data
    except requests.exceptions.HTTPError as err:
        print(f"HTTP error occurred: {err}")
        return None
    except Exception as e:
        print(f"An error occurred: {e}")
        return None


def search_repos(query, per_page=5000):
    url = f'https://api.github.com/search/repositories?q={query}&sort=stars&order=desc&per_page={per_page}'
    try:
        response = requests.get(url, headers=HEADERS)
        response.raise_for_status()
        return response.json()['items']
    except requests.exceptions.HTTPError as err:
        print(f"HTTP error occurred: {err}")
        return []
    except Exception as e:
        print(f"An error occurred: {e}")
        return []

def main():
    technologies = languages_frameworks_systems_sample
    repo_data_list = []

    for tech in technologies:
        print(f"Searching repositories for {tech}...")
        repos = search_repos(tech)
        for repo in repos:
            owner = repo['owner']['login']
            repo_name = repo['name']
            print(f"Fetching data for {owner}/{repo_name}...")
            repo_data = get_repo_data(owner, repo_name, tech)
            if repo_data:
                repo_data_list.append(repo_data)
            time.sleep(1)  # To avoid hitting the rate limit

    df = pd.DataFrame(repo_data_list)
    df.to_csv('languages_frameworks_sample.csv', index=False)
    print(df)

if __name__ == "__main__":
    main()
```
